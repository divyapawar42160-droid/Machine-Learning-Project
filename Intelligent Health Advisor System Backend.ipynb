```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']

# ecoding prognonsis
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
    
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
```

# Decision Tree

```python
dt_classifier = DecisionTreeClassifier(max_depth=100, random_state=42)
dt_classifier.fit(X_train, y_train)
y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
dt_classifier = DecisionTreeClassifier(max_depth=50, random_state=42)
dt_classifier.fit(X_train, y_train)
y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'dt_classifier', 'X', 'y' are already defined

train_sizes, train_scores, test_scores = learning_curve(
    dt_classifier, X, y, cv=5, n_jobs=-1,
    train_sizes=np.linspace(0.05, 1.0, 20)  # Increase number of training sizes
)

# Mean scores and standard deviations
train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)
train_std = train_scores.std(axis=1)
test_std = test_scores.std(axis=1)

# Plot the curves
plt.plot(train_sizes, train_mean, label='Training score', color='blue', linestyle='-', linewidth=2)
plt.plot(train_sizes, test_mean, label='Test score', color='green', linestyle='-', linewidth=2)
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.2, color='blue')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.2, color='green')
plt.xlabel('Training Size')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Learning Curves (Overfitting Check)')
plt.grid(True)
plt.show()
```
